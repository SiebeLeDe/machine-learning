**How to define a function**

The syntax of a function definition in python looks like:
```python
def function_name(arguments):
    """Optional: Docstring to describe the function."""
    # Code block (indented)
    return result  # Optional (only if you want to return something)
```
- ```def``` keyword is used to define a function
- **function_name**: Name of the function (use descriptive names!).
- **arguments**: Variables that the function will use (can also be empty if no arguments are needed). If the argument value is not specified then you have to always provide the argument. If the argument has a default value  you can call the function without that argument and the program will just use the default value of the parameter. Arguments with default values are called *optional*. See examples below.
- **docstring**: String in triple quotes, which describes what function does. Optional.
- **code block**: The body of the function, **everything should be indented!**
- **return**: A result that the function will return (optional, if not specified the function returns None).

**How to call a function**

If you just define the function no calculation will happen. You need to **call** the function later in the code.
To call a function write ```function(args)``` where args are *actual arguments* from the code.

*Formal arguments* are the arguments that specified in the function definition. 
- They act as placeholders that get values when the function is called.
- They exist only inside the function and are local to it.

*Actual arguments* are the values or variables passed to the function when it is called.

````python
def twice(a): # a is a formal argument
    """Mutiplies a number by two."""
    return 2*a

input_val = 10 
result = twice(input_val) # input_val is an actual argument, a variable that exists in the main code

print(f'result is {result}')
````

If you run the same code and try to print *a* in the main code you'll get an error, as *a* only exists inside the function.

````python
def twice(a): # a is a formal argument
    """Mutiplies a number by two."""
    return 2*a

input_val = 10  
result = twice(input_val) # input_val is an actual argument, variable that exists in the main code

print(f'result is {result}')
print(a)
````

**Required arguments**

These are the arguments, that always must be provided, when you call the function.

````python
def add_numbers(a, b):
    """Adds two numbers."""
    return a + b

result = add_numbers(3, 5)  # Output: 8
print(result)
````

If you try to call function without arguments of with fewer arguments you'll get an error message:

````python
result = add_numbers(3)
print(result)
````

**Optional (Default) Arguments**

If no value is specified, the function will use the default value. Python functions often have multiple optional arguments, you can also give a specific one a value by using ```argument=value``` in the call. The cell below demonstrates this.

````python
def get_cost(n_items, item_price=10.5, n_discount=10):
    """Calculate total cost for a given number of item"""
    if (n_items > n_discount):
        # you get a 10% discount if you buy enough items
        return n_items * item_price * 0.9
    else:
        return n_items * item_price
    
n = 10
cost = get_cost(n)
print(f'Cost of {n} items with the default price is {cost}')

price = 9.8
cost = get_cost(n, price)
print(f'Cost of {n} items with price = {price} is {cost}')

items_for_discount = 5
cost = get_cost(n, n_discount=items_for_discount)
print(f'Cost of {n} items with the default price and a discount is {cost}')
````

**Return multiple values**

A function can return multiple values as a tuple:

````python
def analyze_sentence(sentence):
    """Returns information about a sentence"""
    n_words = len(sentence.split())
    l_sentence = len(sentence)
    return n_words, l_sentence

text = "This is a sentence with ten words and fiftynine characters."
info = analyze_sentence(text)
print(info)

n, l = info
print(n) 
print(l)
````

Or you can call the function like this:

````python
n, l = analyze_sentence(text)
print(n, l) 
````

You can also define a function, which **does not return anything**:

````python
def print_string(input_string): 
    print(input_string)

s = 'Hello, world!' 
print_string(s) 
````

In that case the function does print but returns None:

````python
result = print_string(s)
print(f'Function returns {result}')
````

**Nested functions**

In Python, nested functions are functions defined inside another function. These inner functions can access variables defined in the outer function. The variables that can be accessed is called the scope of a function, so nested functions can access the outer functions variables while non-nested functions can not. 

````python
def cube_surface(x):
    """Demonstrates nested functions by computing the surface area of a cube."""
    
    def surface():
        return x ** 2
        
    return surface() * 6

edge = 3
print(f'The surface area of a cube with an edge of {edge} cm is {cube_surface(edge)} cm\N{SUPERSCRIPT TWO}.')
````

Remark: In this function we used a trick to write a unicode symbol in a text that can sometimes be handy. Another example with nested functions. Note that inner functions ```add()``` and ```subtract```  access variables defined in the outer function.

````python
def calculate(x, y): # Formal arguments
    
    # Inner functions know x and y
    def add():
        return x + y 

    def subtract():
        return x - y

    def multiply():
        return x * y

    return add(),subtract(),multiply()
    
a = 10
b = 5 
add, sub, mul = calculate(a, b)
print(f"Subtraction gives {sub}, adding gives {add}, multiplication gives {mul}")
````
