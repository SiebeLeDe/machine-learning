# Looping, conditionals and lists

Trying out many different values by typing them in by hand is tedious and can be avoided once you know how to make a *loop*. Loops are pieces of code that are executed many times and can be constructed in all programming languages. In Python there are various ways to set up a loop, common feature is that there is one variable that changes in a systematic way. 

Let's start the example, where we want to calculate a sum over many number from 1 to 100:

Read Loops_summary.ipynb to understand how the ```range()``` function works.

````python
sum_numbers = 0
for i in range(90, 111):
    sum_numbers += i
print(f'The sum of numbers is {sum_numbers}')
````

Let's print each nucleotide of a DNA sequence. We thereby iterate a variable nucl in the string dna_sequence. Because dna_sequence is a string we can iterate over the sequence of characters directly.

````python
dna_sequence = "ATGCGT"

for nucl in dna_sequence:
    print(nucl)
````

When you use a for loop in the example above you get only the character, but not its position in the string. 
The function ```enumerate``` is very useful when you also need the position. It takes an iteratable (string in our case) as an argument and returns a tuple containing the index and the value from the iterable.

````python
dna_sequence = "ATGCGT"

for i, nucl in enumerate(dna_sequence):
    print(f'The nucleotide {nucl} is at position {i}')
````

Loops are very often nested, meaning that you iterate again inside a loop. Understanding how this works is very important as you will encounter this a lot in practical applications of programming. In the example below we use two loops to print a triangle with numbers.

````python
rows = 5
for i in range(1, rows + 1):  # Outer loop for number of rows
    for j in range(1, i):  # Inner loop for printing a triangle in each row
        print(j, end = '')  # Print a 
    print()  # New line after each row
````

Note the in the cell above we use additional argument in ```print()```.
The end parameter in the print() function specifies what character (or string) to print at the end of the output. By default, ```end = "\n"```, which means a newline is printed after the output.
You can change this to any string, such as a space, comma, or even an empty string, to modify how the output is printed.

# Conditional statements

**Read first the summary notebook about conditional statements: Conditionals.ipynb**

The if statement is used for conditional execution of the code. Like loops, these are ended with a colon `:` and indentation is used to indicate which piece of code is used if the *conditional expression* evaluates to `True`. We can also provide an `else:` block that is only executed if the *conditional expression* evaluates to `False`.

- **if Statement** Executes a block of code if a condition is True.
```python
if condition:
    # Code to execute if condition is True
```
- **if-else Statement** Executes one block of code if the condition is True, otherwise executes another block.
```python
if condition:
    # Code to execute if condition is True
else:
    # Code to execute if condition is False
```

- **if-elif-else Statement** Checks multiple conditions. Executes the block of the first True condition. You can add as much **elif** statements as you want.

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition2 is True
else:
    # Code to execute if none of the conditions are True
```

Note that operator ```=``` is for assignment, while ```==``` is for comparison.

```python
if x = 10:  # Incorrect
if x == 10:  # Correct
```

The code below takes the list of student grades and categorize the grades into three groups:
   - Failing: Grades less than 50
   - Passing: Grades from 50 to 74 (inclusive)
   - Excellent: Grades 75 and above

Also it determines the top 5 grades to gain insight into the highest-performing students.

**Exercise:** *Modify the code to get the 5 lowest grades.*

````python
# List of student grades
grades = [45, 76, 89, 55, 90, 34, 73, 100, 67, 49, 83, 42, 78]

# Categorize grades
failing = []
passing = []
excellent = []

for grade in grades:
    if grade < 50:
        failing.append(grade)
    elif grade < 75:
        passing.append(grade)
    else:
        excellent.append(grade)

# Determine the top 5 grades
sorted_grades = sorted(grades, reverse=True)
top_five_grades = sorted_grades[:5]

# Print the analysis
print("Number of failing grades:", len(failing))
print("Failing grades:", failing)

print("Number of passing grades:", len(passing))
print("Passing grades:", passing)

print("Number of excellent grades:", len(excellent))
print("Excellent grades:", excellent)

print("Top five grades:", top_five_grades)
````

Sorting of numbers is easy, but when sorting strings you need to consider also the case. The example below combines two parts of the oldschool reading method and then sorts the combined list. Note that this leads to an order in which all the person names are listed first because they start with an uppercase character. 

![Leesplankje](Leesplankje.png)

````python
leesplankje = ['aap','noot']
leesplankje += ['Mies','Wim','zus','Jet', 'Teun', 'vuur', 'Gijs','lam','Kees','bok','weide']
leesplankje.sort()
print(leesplankje)
````

Python offers many approaches to get a proper sorting on alphabet while ignoring the case, one that demonstrates the use of the ```insert``` method is given below. It uses the fact that the first two words are already in correct order and then places the one to be added at the proper place. This approach is certainly not the best (most compact) approach but used here to demonstrate the use of various language elements that we encountered before. Take a moment to analyze this code and make sure you understand what happens.

**Exercise:** *Place the comment symbol ```#``` before ```break``` to deactivate this statement and see that the cell never completes when you run it.*

Removing the break creates an endless loop that should be ended by choosing *interrupt kernel* from the Kernel menu.

````python
leesplankje = ['aap','noot']
for new_word in ['Mies','Wim','zus','Jet', 'Teun', 'vuur', 'Gijs','lam','Kees','bok']:
    for index, word in enumerate(leesplankje):
        if word.lower() > new_word.lower():
            leesplankje.insert(index,new_word)
            break
    if word.lower() < new_word.lower():
        leesplankje.append(new_word)
print(leesplankje)
````
