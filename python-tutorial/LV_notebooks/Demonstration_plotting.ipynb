## Demonstration of plotting

Some demonstrations of plotting using matplotlib.

Let's first plot a sine function over the interval $[0,5\pi]$, taking 500 points to get a smooth curve.

```python
import matplotlib.pyplot as plt
import numpy as np

def function_1(x,a=1):
    return np.sin(a*x)

x = np.linspace(0,5*np.pi,500) # Using the linspace function to get x-values
y_1def = function_1(x)         # Call the function to get the standard y-values
y_1mod = function_1(x,3)       # Call the function again but now with modified parameters

plt.plot(x,y_1def,'red')        # Plot function_1 with default parameters
plt.plot(x,y_1mod,'blue')       # Plot function_1 with modified parameters
plt.show()                      # Show the plot
```

We can also combine scatter and line plots. An example with a noisy and a smooth third order polynomial.

```python
import numpy as np

def function_2(x,a=0.0,b=1.0,c=0.0,d=0.0,noise=0.0):
    '''
    Simulate noise on a third order polynomial.
    '''
    y = a + b*x + c*x**2 + d*x**3 + noise*(0.5-np.random.sample(x.shape))
    return y
    
x = np.linspace(-5,5,200)
y1_smooth = function_2(x,2.0,1.0,-1.0,0.5)
y1_noisy  = function_2(x,2.0,1.0,-1.0,0.5,noise=5)
y2_smooth = function_2(x,-28.0,c=4.0)
y2_noisy  = function_2(x,-28.0,c=4.0,noise=7)

plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y1_smooth,c='red')
plt.scatter(x,y1_noisy,c='darkred',marker='+',s=10)
plt.plot(x,y2_smooth,c='blue')
plt.scatter(x,y2_noisy,c='darkgreen',marker='^',s=10)

plt.title('Drawing polynomials through noisy data')
plt.show()
```

We can define logical filters to restrict ourselves to 
data that fulfills certain criteria

```python
# Read dataset by Shibata, Suzuki, Mizoguchi (DOI:10.17632/yz5rrmvrgd.1)
file = 'diatomic_df.csv'
data = np.loadtxt(file, skiprows=1, usecols=(11,14),max_rows=1430,delimiter=',')

# Bond lengths shorter than 3 Å
filter_1 = data[:,0] < 3.
# Fermi energy below -7 eV
filter_2 = data[:,1] < -7.
# Both criteria fulfilled ?
filter = np.logical_and(filter_1,filter_2)

# Show the filters
print(filter_1[:10])
print(filter_2[:10])
print(filter[:10])

plt.xlabel('bond distance in Å')
plt.ylabel('Fermi energy in eV')
filtered_data = data[filter]
plt.scatter(filtered_data[:,0],filtered_data[:,1],
            c='grey',marker='s',s=20)
plt.show()
```

```python
"""
@author: tteodoro

Plots the concentration of a reaction with reactant A forming B and C vs time.
"""

############## Modules ###############
import matplotlib.pyplot as plt
import numpy as np

##############  Input  ###############
n_solut = 3  # How many solutions to print
box_len = 2  # Length of the box
spacing = 0  # Spacing of the functions in the plot

############# Functions ##############
def energy(n, L):
    """
    Calculates the energy of the particle.
    """
    return n**2/(8*L**2)

def psi(n, L, x):
    """
    Wave function value in (a) given point(s) along the length of the box.
    """
    return  np.sqrt(2/L)*np.sin(n*np.pi*x/L)


######## Running and output #########

# List with x values for calculating psi and plotting
x = np.linspace(0,box_len,100)

# Plot text
plt.title('Wave functions of particle in a box of length {}'.format(box_len))
plt.xlabel('Position in box')
plt.ylabel('Wave function')

# Plot bounding box
plt.xlim(0.0, box_len)

# Plot the wave functions
for n in range(1, n_solut+1):
    y = psi(n,box_len,x)
    plt.plot(x, y,label=f'n={n:2}, E={energy(n,box_len):6.3f}')

plt.legend()
plt.show()
```

```python
"""
@author: tteodoro

Plots the concentration of a reaction with reactant A forming B and C vs time.
"""

############## Modules ###############
import matplotlib.pyplot as plt
import numpy as np


##############  Input  ###############
# First-order rate constants
k1   = 300  # s^{-1}
k2   = 100  # s^{-1}
# Initial concentration of A
A0   = 2.0  # mol dm^{-3}
# Maximum time in the plotting
tmax = 15   # ms 


############# Functions ##############
def concentrations(A0, k1, k2, t):
    """Calculates the concentration of the reactant and products at time t."""
    # Converts time from s to ms
    t = t * 1e-3
    # Calculates concentrations
    A = A0 * np.exp(-(k1+k2)*t)
    B = (k1/(k1+k2)) * A0 * (1 - np.exp(-(k1+k2)*t))
    C = (k2/(k1+k2)) * A0 * (1 - np.exp(-(k1+k2)*t))
    
    return A, B, C


############# Running  ###############
# List with x values (time)
x_axis = [x for x in np.arange(0.0,tmax,0.1)]
# Calculates concentrations over time
concs  = [concentrations(A0, k1, k2, t) for t in x_axis]
# Unzip concs into individual molecular concentrations
concA, concB, concC = zip(*concs) # Not really necessary if labels not required


############# Output ################
# Plot text
plt.title('Evolution of molecular concentration over time')
plt.xlabel('Time (ms)')
plt.ylabel('Concentration (mol / dm\u00b3)')

# Plot configuration
plt.xlim(0.0, tmax)

# Plot to the terminal
plt.plot(x_axis, concA, label='[A]')
plt.plot(x_axis, concB, label='[B]')
plt.plot(x_axis, concC, label='[C]')
plt.legend()
plt.show()
```
