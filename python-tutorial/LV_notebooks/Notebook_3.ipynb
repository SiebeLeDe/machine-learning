## Organizing numerical data with numpy

We already encountered the concept of a list to organize data, but for numerical data it is more convenient and much faster to use the `numpy` library that allows us to define vectors and matrices and do mathematical operations on lot's of data at the same time. In the simplest form it can do operations that you may do in an Excel sheet, but it can do much more. Let's start with a simple example of determining a grade for students that have each made 3 assignments with varying weights. These grades are stored in a comma-separated-file (.csf) that can be read into Python using numpy. You can check the contents of this file by opening it with (for instance) excel.

````python
import numpy as np

file = 'ExamResults.csv'

midterm = np.loadtxt(file, skiprows=1, usecols=1, delimiter=',')
practical = np.loadtxt(file, skiprows=1, usecols=2, delimiter=',')
exam = np.loadtxt(file, skiprows=1, usecols=3, delimiter=',')
final = 0.2 * midterm + 0.3 * practical + 0.5 * exam
print(final)
````

You can check the type of the loaded grades and see that it is a new type (more formal: a new class) of data:

````python
print(f'Type of the loaded grades is {type(midterm)}')
print(f'Type of the loaded grades is {type(final)}')
````

The lists of numbers that are read in (midterm, practical and exam) or automatically created (final) are called `numpy arrays`. In this case they are all 1-dimensional. **Read Arrays_summary.ipynb to get examples how to work with arrays.**

A few things to note in the above:

- When importing you can rename the library, it is thereby conventional to shorten numpy to np. Use this convention also in your notebooks.
- The function `np.loadtxt` can read data stored in a textfile and has optional arguments to skip one or more lines (`skiprows`) and/or to select the desired column (`usecols`). The default is to work with data separated by whitespaces, so we indicate that our data is comma-separated instead (`delimiter=','`).
- Documentation of np.loadtxt is found at https://numpy.org/doc/stable/reference/generated/numpy.loadtxt.html. 

You see that all individual marks are multiplied and added to obtain the final list of grades. Exactly the same we could have done in a loop (see below), but it is usually faster and easier to use numpy. 

````python
# You need to use a loop to get grades using lists

n_students = midterm.size

final_list = [0] * n_students
midterm_list = list(midterm)
practical_list = list(practical)
exam_list = list(exam)

for i in range(0, n_students):
    final_list[i] = 0.2 * midterm_list[i] + 0.3 * practical_list[i] + 0.5 * exam_list[i]
    
print('Differences between the two ways of calculating',final - final_list)
````

Note that in the cell below we use ```list(midterm)``` to convert numpy arrays to list (see the summary notebook). Since we were working with lists we created a list of zeros using ```[0] * n_students```. 

When working with ```numpy```, it is simple to create an array of zeros with the built-in method ```np.zeros``` (see the summary notebook for more ```numpy``` built-in functions. For this we did need to know the size of the array which we could get from the ```attribute``` `midterm.size`. Size also works also for multidimensional arrays, where it will give you the total number of elements. 

**Multidimensional arrays**

We will demonstrate the use of a multidimensional array by reading in the data once more, but now all at once. Note again that Python counts from 0, so that for getting the marks of the second student you need index 1. Note also that for usecols we now gave three numbers in parenthesis, such a list of values is called a *tuple* and differs from an ordinary list by being *unmutable*, meaning we can not add or change elements. They are used to pass several values at once to a function or when the result of a function logically consists of multiple values.

The colons `:` indicate that an entire row or column of the array is used. You can also specify a range, for instance `0:8` if you want the first 8 numbers (starting with index 0 and having index 7 as the last). 

````python
marks = np.loadtxt(file, skiprows=1, usecols=(1, 2, 3), delimiter=',')

print(f'Size of the marks array is {marks.size}') # For multidimensional arrays gives total number of elements
print(f'Shape of the marks array is {marks.shape}') # 2d array 24 x 3
print()

check_final = 0.2 * marks[:, 0] + 0.3 * marks[:, 1] + 0.5 * marks[:, 2]
print(f'All marks of student number 2: {marks[1, :]}')
print(f'The results for the practical of student 5: {marks[4, 1]}')
print(f'Difference between two ways of calculating: {check_final - final}')
````

**The linspace function in numpy**

An often-used numpy function is `np.linspace` to generate an array with evenly spaced numbers over a specified range. It is especially useful when you need a fixed number of values within a range, such as creating grids, defining intervals for plots, or setting up arrays for numerical simulations.

```python
np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
```

Documentation on the linspace function is available at https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

We end this introduction to the use of numpy arrays by looking at the possibility to fill arrays with random numbers, which can be useful if we want to simulate processes governed by chance. An example is a lottery in which a number of prizes is distributed randomly over a large number of people.

Suppose we have a lottery in which there are 100 prizes of 5€, 30 prizes of 50€ and the big prize of €1000 euro. There are 5000 tickets available that cost 1€ each that may or not be sold. As the organizer of the lottery you want to compute the average profit you can expect, knowing that between 3000 and 4800 of the tickets will be sold for each lottery. We can try to compute this profit using statistical methods, but it is much quicker to write a small Python script that simulates a large number of draws and computes the profit for each draw. This program is given below and serves to show a number of useful functions of numpy. Remember for the third line the possibility to concatenate lists that is used here to fill the array with prizes. Note also the use of comments to explain what each line of code is doing. Try running the cell a few times and see how the maximum and minimum profit vary from simulation to simulation, whereas the average fluctuates much less.

````python
import random                                     # For getting a single random number we use the random library

drawings = 10000                                  # How many drawings we want to simulate
profit = np.empty((drawings))                       # We create an array to store the profit made in each drawing
tickets = np.array(4869 * [0] + 100 * [5] + 30 * [50] + [1000]) # The array with all prizes (non-winning tickets have prize 0)
for lottery in range(drawings):                   # Loop over all drawings
    n_tickets = int(random.uniform(3000, 4800))   # Number of tickets sold in this drawing
    # Make a new array with all tickets sold in this drawing
    tickets_sold = np.random.choice(tickets, n_tickets, replace=False)   
    prize_money = np.sum(tickets_sold)            # The sum function adds all elements in the array
    profit[lottery] = n_tickets - prize_money     # Income is n_tickets terms 1 euro, expense is the prize_money
    
print (f'The maximum profit on a single drawing was {np.max(profit)} euro.')
print (f'The minimum profit on a single drawing was {np.min(profit)} euro.')
print (f'The average profit is {np.average(profit):.2f} euro.')
````
