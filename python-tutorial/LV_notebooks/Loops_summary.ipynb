# Loops


**The for loop**

A ```for``` loop in Python is used to repeat a block of code for each item in a sequence (like a string or range of numbers).

Usage definition:

```python
for variable in sequence:
    # Code block to execute
```

You can iterate through the charachters in string.

````python
# A string to loop over
word = "Python"

# Using a for loop
for char in word:
    print(char)  # Prints each character on a new line
````

The ```enumerate()``` function in Python is used to iterate over an iterable (e.g. a string) while keeping track of the index of the current item. It returns a tuple containing the index and the value from the iterable, making it very useful when you need both the item and its position.

````python
for index, char in enumerate('Python'):
    print(f"Index {index}: {char}")
````

**Range function**. The range() function generates a sequence of numbers from a start value to a stop value (not included) with a given step.

```python
range(start, stop, step)
```

The default values are:
- Start = 0
- Step = 1

So calling the function range as:
```python
range(stop)
```
will be equal to calling it with default values
```python
range(0, stop, 1)
```

To see the sequence produced by range() we need to use a for loop:

````python
# Generate numbers 0, 1, 2, 3, 4

for i in range(5):
    print(i)
````

````python
# Generate odd numbers

for i in range(1, 10, 2):
    print(i)
````

````python
# Generate number 1, 2, 3

for i in range(1, 4):
    print(i)
````

You can also use a negative step value to count backwards:

````python
# Loops from 10 to 2, stepping down by 2

for i in range(10, 0, -2): 
    print(i)
````

**Nested loops**

A nested loop consists of an outer loop and one or more inner loops. For each value of the outer loop variable, the inner loop runs completely.

````python
for i in range(3):  # Outer loop
    for j in range(4):  # Inner loop
        print(f"i = {i}, j = {j}")
````
