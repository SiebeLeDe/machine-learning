# Variables, numerical data types and assignments in Python

In Python, variables are used to store data values, and they are created the moment you assign a value using the $=$ operator.

Variable names should be descriptive, follow naming conventions (e.g., ```reaction_rate```), and must start with a letter or an underscore, but cannot start with a number or contain special characters.

Variables can hold different types of data. In this summary we focus on the three main numerical datatypes:
- Integer (```int```): Whole numbers (e.g. 5, -121, 0)
- Floating-point (```float```): Numbers with a decimal point (e.g. 3.14, -0.5, 2.72)
- Complex (```complex```): Numbers in the form a + bj, where a is the real part and b is the imaginary part (e.g. 2.1 + 5.6j, 0.5 - 0.23j)

You can get and print the type of your variable with the built-in function ```type()```:

````python
x = -121             # Integer
print(type(x)) 

y = 3.14             # Float
print(type(y)) 

z = 0.5 - 0.23j      # Complex
print(type(z)) 
````

---
**Integer (int)**

Operations:
- Addition: +
- Subtraction: -
- Multiplication: *
- Floor division: //
- Modulus: %
- Exponentiation: **

````python
# Integers
a = 10
b = 3

# Basic operations
print("a =", a, "b =", b)
print("Addition (a + b) =", a + b)       
print("Subtraction (a - b) =", a - b)    
print("Multiplication (a * b) =", a * b) 
print('Divition (a / b) =', a / b) # Note that regular division returns a float
print("Division (integer) (a // b) =", a // b)  # This is an integer division
print("Remainder (a % b) =", a % b)      
print("Exponentiation (a ** b) =", a ** b) 
````

---
**Floating-Point Numbers (float)**

Operations:
- Similar to integers, but division (/) always produces a float.
- Division (//) truncates the result to the nearest integer less than or equal to the division result. When applied to floats, the result is still a float.

````python
# Floats
x = 5.5
y = -2.1

# Basic operations
print("x =", x, "y =", y)
print("Addition:", x + y)         
print("Subtraction:", x - y)      
print("Multiplication:", x * y)   
print("Division:", x / y)         
print("Division:", x // y)         # the result is still a float
print("Exponentiation:", x ** y)  
````

---
**Complex Numbers (complex)**

Operations:
- Addition, subtraction, multiplication, and division work as expected.
- Use .real and .imag attributes to access the real and imaginary parts.
- Use the abs() function to compute the magnitude.

````python
# Complex numbers
z1 = 2 + 3j
z2 = 1 - 4j

# Basic operations
print("z1 =", z1, "z2 =", z2)
print("Addition:", z1 + z2)        
print("Subtraction:", z1 - z2)      
print("Multiplication:", z1 * z2)  
print("Division:", z1 / z2)         

# Accessing real and imaginary parts
print("Real part:", z1.real)        
print("Imaginary part:", z1.imag)

# Magnitude of a complex number
print("Magnitude of z1:", abs(z1))  
````

````python
# You can see that // does not work with complex numbers
print('Floor division:', z1 // z2)
````

````python
# Also % does not work with complex numbers.
z1 % z2
````

**Augmented assigment statements**

Augmented assignment statements are a shorthand way of updating the value of a variable. 
Instead of writing the variable name twice, you can use an augmented assignment operator to simplify the code.

Here are some common augmented assignment operators and their usage:

- Addition (+=)
- Subtraction (-=)
- Multiplication (*=)
- Division (/=)
etc

To read more search for "Augmented assignment statements" with Google or another search engine. An easy intro is given at https://www.geeksforgeeks.org/augmented-assignment-operators-in-python/, the more complete (but quite formal) definition is found in the offical documentation at https://docs.python.org/".

````python
# Addition statement
x = 5
x += 3  # Equivalent to x = x + 3
print('The addition result is:', x)  

# Subtraction statement
y = 10
y -= 4  # Equivalent to y = y - 4
print('The subtraction result is:', y)  

# Multiplication statement
z = 7
z *= 2  # Equivalent to z = z * 2
print('The multplication result is:', z) 

# Division statement
a = 20
a /= 5  # Equivalent to a = a / 5
print('The division result is:', a) 
````

**A wide range of mathematical functions is implemented in the math module**

- You can import the whole module by writing: ```import math```

Then you can call all functions that are contained inside the module as: ```math.cos()```, ```math.sin()```, ```math.acos()```, etc. 

For the information about the math functions see the documentation at https://docs.python.org/3/library/math.html

- You can also import only some specific functions from the module as: ```from math import cos, sin```
  
If you import specific functions from a library, your notebook only knows these and not the other functions available in that library.

-------------------------------------------------------------
**The math module also contains some useful constants**

- **math.pi**: The value of π.

Example: math.pi → 3.141592653589793

- **math.e**: The base of the natural logarithm (e ≈ 2.718).

Example: math.e → 2.718281828459045

````python
#Let's print the values of these constants

from math import pi, e

print('π is:', pi)
print('e is:', e)
````

-------------------------------------------------------------
**Trigonometric functions:**

Note that all these functions return a value with float type!

- **math.sin(x)**: Returns the sine of x (in radians).

Example: math.sin(math.pi / 2) → 1.0

- **math.cos(x)**: Returns the cosine of x (in radians).

Example: math.cos(0) → 1.0

- **math.tan(x)**: Returns the tangent of x (in radians).

Example: math.tan(math.pi / 4) → 1.0

````python
#Let's use the functions with a value of pi/2

from math import sin, cos, tan, pi

print('The sine of pi/2:', sin(pi/2))
print('The cosine of pi/2:', cos(pi/2))
print('The tangens of pi/2:', tan(pi/2))
````

-------------------------------------------------------------
**More trigonometric functions:**

Note that all these functions return a value with float type!

- **math.asin(x)**:Returns the arcsine of x, in radians.
  
Example: math.asin(1) → 1.5707963267948966 (π/2)

- **math.acos(x)**: Returns the arccosine of x, in radians.
  
Example: math.acos(1) → 0.0

- **math.atan(x)**: Returns the arctangent of x, in radians.

Example: math.atan(1) → 0.7853981633974483 (π/4)

````python
#Let's use the functions with a value of 1

from math import asin, acos, atan

print('The arc sine of 1 is:', asin(1))
print('The arc cosine of 1 is:', acos(1))
print('The arc tangent of 1 is:', atan(1))
````

-------------------------------------------------------------
**More math functions:**

- **math.sqrt(x)**: Computes the square root of x

Example: math.sqrt(16) → 4.0

- **math.factorial(x)**: Returns the factorial of x (x!) (only for integers).

Example: math.factorial(5) → 120

````python
from math import sqrt, factorial, fabs

print('The square root of 4:', sqrt(4))
print('The factorial of 4:', factorial(4))
````

-------------------------------------------------------------
**Logarithmic and Exponential functions:**

 - **math.log(x, base)**: Returns the logarithm of x to the specified base. 

If no base is given, computes the natural logarithm (lnx)
Example: math.log(8, 2) → 3.0

 - **math.log10(x)**: Returns the base-10 logarithm of x.

Example: math.log10(1000) → 3.0

- **math.exp(x)**:
Returns $e^x$ (exponential of x)
  
Example: math.exp(2) → 7.389056098930649

````python
from math import log, log10, exp

print('The base-2 logarithm', log(8, 2))
print('The base-10 logarithm:', log10(1000))
print('Also the base-10 logarithm:', log(1000, 10))
print('Natural logarithm:', log(e ** 2))
print('Exponent:', exp(2))
````

For working with complex numbers use the **cmath** module instead of math:

````python
import cmath

print('Exponent of the imaginary number i:', cmath.exp(1j))
print('Logarithm of the imaginary number i:', cmath.log(1j))
print('Square of the imaginary number i:', cmath.sqrt(1j))
````

-------------------------------------------------------------
**If you want to check the closeness of two values**
    
- **math.isclose(a, b, abs_tol=1e-9)**: Checks if two values are close to each other within a absolute tolerance.

Example: math.isclose(0.1 + 0.2, 0.3) → True

````python
import math 

print(math.isclose(0.1 + 0.2, 0.3))
print(math.isclose(0.1 + 0.2, 0.35))
````

The result is a boolean, a data type that can only have the values True or False.
