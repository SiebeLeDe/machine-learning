# Part 1. Strings

Textual data in Python is handled with ```str``` objects, or strings. Strings are immutable sequences. String literals are written in a variety of ways:

- Single quotes: 'allows embedded "double" quotes'
- Double quotes: "allows embedded 'single' quotes"
- Triple quoted: '''Three single quotes''', """Three double quotes"""

Triple quoted strings may span multiple lines.

````python
my_string = 'Hello, World!'
another_string = "Python is awesome!"
long_string = '''
that can be any long string
with multiple lines
'''

print(my_string)
print(another_string)
print(long_string)

# You can check the type of your variable by using type() function:
print(type(my_string))
print(type(another_string))
print(type(long_string))
````

**Joining strings:** You can join two or more strings using the ```+``` operator:

````python
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2  # Concatenating with a space in between
print('Result of joining strings:', result)
````

**Repeat strings**: You can repeat a string using the ```*``` operator.

````python
word = "Python"
repeated_word = word * 3  # Repeating the string 3 times
print('Repeated string:', repeated_word) 
````

**String length:** You can get a length of the string using built-in ```len()``` function:

````python
my_string = "Hello"
length = len(my_string)
print(length)
````

**Accessing characters:** 

Strings in Python are indexed starting from $0$. You can access individual characters by their index.

You can use negative indices, to access characters from the last character and so on.

For a string with N characters:  
- With positive indexing the indices run from $0$ to $N-1$.   
- With negative indexing the indices run from $-1$ ($N-1$ in positive indexing) to $-N$ ($0$ in positive indexing).  

````python
my_string = "Hello"
l = len(my_string)

first_char = my_string[0]  # Get the first character
first_char_neg = my_string[-l] # Also get the first character with negative indexing

last_char = my_string[-1]  # Get the last character
last_char_neg = my_string[l-1] # Also get the last character

#You can check that the characters are the same
print('Positive and negative indices return the same result:', first_char == first_char_neg)

print('String:', my_string)
print('First character:', first_char)
print('Last character:', last_char)
````

**Slicing a string:** You can extract a substring using slicing. Note that the left index is included, the right index is excluded.

````python
my_string = "Hello, World!"
substring = my_string[7:12]  # Extracts "World"
print(substring)
````

**Checking String Membership:** You can check if a substring exists within a string using **in** keyword.

````python
my_string = "Hello, World!"
print("Hello is in my string:", 'Hello' in my_string) 
print("hello is in my string:", 'hello' in my_string) 
````

**Changing the case:** You can change the case of a string using methods like ```.lower()``` and ```.upper()```:

Note that before we used functions like ```len(argument)```, where we put the arguments in brackets. Many functions take different kinds of arguments and are therefore not tied to a particular class of objects. 

A **method** is a function that belongs to a specific class of objects and is called as ```object.method()```, i.e. having a dot separate the method from the object. Like functions, a method may have one or more arguments.

````python
my_string = "Hello, World!"
print(my_string.lower()) 
print(my_string.upper()) 
````

**Replacing substrings:** You can replace parts of a string using the .replace() method. 

````python
my_string = "Hello, World!"
new_string = my_string.replace("World", "Python")
print('my string:', my_string)
print('new string:', new_string) 
````

**String splitting:** You can split a string into a list using the ```.split()``` method. By default, it splits by whitespace symbols (space, newline, tab). This method returns a list of strings.

````python
my_string = 'Hello world, python'

split_string_default = my_string.split() # Split by default (space here)
split_string = my_string.split(", ")  # Split by comma and space

print(split_string_default)
print(split_string) 
````

**Counting substrings**: You can use count() **method** to count the number of occurrences of a substring in a string:

````python
text = "Python is fun. Python is easy to learn."
print('The word Python appears in the string', text.count('Python'), 'times')
````

# Part 2. Formatting text with f-strings

A formatted string literal or **f-string** is a string literal that is prefixed with 'f' or 'F'. These strings may contain replacement fields, which are expressions delimited by curly braces {}. While other string literals always have a constant value, formatted strings are really expressions evaluated at run time.

Example:

    f'Something is written inside {x}', where x is the variable that you want to substitute with it's value

- f-strings allow to **embed variables directly into a string**:

````python
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
````

- f-strings allow to **perform calculations directly inside the string:**

````python
a = 5
b = 3
print(f"The sum of {a} and {b} is {a + b}.")
````

- f-strings allow to **format numbers:**

````python
pi = 3.14159
print(f"Pi rounded to 2 decimal places is {pi:.2f}.")
````

- f-strings allow to **align the output:**

Right-aligned: The value is aligned to the right within a field of 5 spaces.  
Left-aligned: The value is aligned to the left within a field of 5 spaces.  
Center-aligned: The value is centered within a field of 5 spaces.  

````python
value = 42
print(f"Value right-aligned: {value:>5}")
print(f"Value left-aligned: {value:<5}")
print(f"Value center-aligned: {value:^5}")
````

- You can use **conditional expressions directly in the f-string:**

````python
score = 85
print(f"You {'passed' if score > 50 else 'failed'} the exam.")
````

# Part 3. Input data

Until this point we defined the variables' values in the cells. However, in a lot of cases you want to get variables from the user or from external file.

**Input function**

The input() function in Python is used to take user input from the standard input (typically the keyboard). It reads a line of text entered by the user and returns it as a string. The program waits until the user provides input and presses Enter.

Run the cell below and see how input() function works.

````python
print("Please enter your full name")
name = input()
print(f"Nice to meet you, {name}")
````

**Read from a file**

Reading from a file in Python is a common task that allows you to access and process data stored in files. 

- Open the File: Use the ```open()``` function to open the file. This function returns a file object.
- Read the File: Use methods like ```read()```, ```readline()```, or ```readlines()``` to read the file's content.
- Close the File: Always close the file after reading to free up system resources.

````python
file_name = 'titin.txt'

file = open(file_name, 'r') # Open the file in read mode ('r')
molecule_name = file.read()
file.close()

print(f'The full name of titin is {len(molecule_name)} characters long.')
print(f'In titin there are {molecule_name.count("asparaginyl")} asparagins.')
````
