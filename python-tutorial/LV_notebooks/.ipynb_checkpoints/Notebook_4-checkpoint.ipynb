## Making your own functions

As you have seen there are already very many functions available that you can use but as you coding skills grow, it will be useful to know how you can define these yourself. 

You probably already saw some examples when looking up documentation and may think that this is complicated and only doable for expert programmers. This is not the case, the problem is simply that by using Google you may hit complicated functions that use concepts you are not yet familiar with. For this reason we will go back to simple variables and only use the math library to demonstrate how making and using a function works. We will pick an example that also demonstrates again the `if` statement which is besides *loops* the most-used elementary building block found in any programming language.

To illustrate making your own function we consider the $sinc(x)=\frac{sin(x)}{x}$ mathematical function that has a well-defined limit $sincx(0)=1$ for x=0, but for which literal translation into Python gives an error:

````python
from math import sin
x = 0
sinc = sin(x)/x
````

We can remedy this by introducing an `if` statement (see Conditionals.ipynb for more information on the if statement). 

````python
from math import sin
import numpy as np

print('       x   sin(x)  sinc(x)')
x_vals = np.linspace(-0.10, 0.10, 5)
for x in x_vals:
    if (abs(x) > 1.e-8):
        print (f'{x:8.2f} {sin(x):8.5f} {sin(x)/x:8.5f}')
    else:
        print (f'{x:8.2f} {sin(x):8.5f} {1:8.5f}')
````

You see that we can now also include x-values close or equal to zero and get the correct table of function values, but the code has become longer and contains two lines that are very similar. 

If we not just want to print sinc(x) but also want to use this in calculations we would have to introduce similar if statements in all cases that x can become so small that an error would occur in the division. To keep code compact and readable, functions can be introduced to perform a recurring task. For the sincx(x) function this works as follows.

````python
from math import sin

def sinc(x, degrees=False):
    ''' 
    Evaluates the mathematical function sin(x)/x.
    x to be given in radians(default) or degrees.
    '''
    if degrees:
        from math import radians
        x = radians(x)
    if (abs(x) > 1.e-8):
        return sin(x) / x
    else:
        return 1.0
    
print('       x   sin(x)  sinc(x)')
x_vals = np.linspace(-0.10, 0.10, 5)

for x in x_vals:
    print (f'{x:8.2f} {sin(x):8.5f} {sinc(x):8.5f}') # This is where we call the function sinc(x), which we just defined above
````

**Let's take a closer look at how functions are defined.**

- The first line starts with `def` followed by the name of the function ```sinc``` and in parentheses the *arguments* x and degrees.
- The argument ```degrees``` is *optional* because a *default* value of `False` is specified, the argument ```x``` has no default and should always be specified.


**Read more about functions in Functions_summary.ipynb.**

A function can have as many arguments as you like and having optional arguments is a common way to allow for special cases or finetuning of the way the function works. You will encounter many examples of such arguments when we will discuss plotting next time as they are then used to customize the way the plot looks like.

- The lines surrounded by three quotes provide information about the function.

It is good practice to always write these so-called documentation strings (docstrings), because you can easily look them up in Python. In Jupyter notebook this is done by selecting the function inside a code block and pressing shift-tab.

**Put your cursor** on the place where the ```sinc``` or ```sin``` function is called and press shift-tab: you will see the docstrings of the selected function.
