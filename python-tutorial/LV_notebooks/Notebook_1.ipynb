## Python as a text processor

Now that you know how to manipulate numerical data we will have a closer look on the reading, modifying and writing texts. We start a simple example of dividing numbers, similar to the one we used before, but now want to have control over how many digits we print. This is most easily done using f-strings (formatted string literals).

```python
distance         = 2633   # travelling distance in km
tank_volume      = 47     # how much liter of fuel will fit in the car
fuel_consumption = 0.074  # liter of fuel needed to drive 1 km
fuel_needed      = distance * fuel_consumption
tank_fills       = fuel_needed / tank_volume

print ("We need",fuel_needed,"liter of fuel so have to refill",tank_fills-1,"times.")
print (f"We need {fuel_needed} liter of fuel so have to refill {tank_fills-1} times.")
```

You see that both lines of code give the same result with a rather ugly way of printing the results. The first number has way too many digits due too rounding errors and the second number does not need digits at all. We can remedy this by using a format specifier:

```python
print (f"We need {fuel_needed:5.1f} liter of fuel so have to refill {tank_fills-1:.0f} times.")
```

In the text above
we used a very simple specifier that just changes the number of digits. You can have much more control, for instance choosing to left or right align numbers or to express them in scientific (exponential) notation if you work with very large or very small values. A full description is found in https://docs.python.org/3/reference/lexical_analysis.html#f-strings. This is a place to bookmark and return to when finetuning the printing of your data. Below we give some examples.

The error messages that you got indicate that Python did not know these mathematical operations. We can `import` them in two ways: explicitly by using `from library import function` or implicitly by using import `math`. The advantage of the first method is that you need not name the library when using the function, the disadvantage is that you need not name the library of the function when using it ! 

With second method that we used in the end with `acos` you will need to specify where the function came from, reducing chances on mistakes due to functions with the same name coming from different libraries. Of course the code is longer in this way, so a programmer who does not like to type will often use the lazy `from math import *` which imports all functions contained in the library. A good compromise is to use `from math import acos, atan`, i.e. just importing the functions that you are going to use. You can then find back where they came from by looking for import statements.

The standard `math` library contains the most used mathematical functions and we refer to https://docs.python.org/3/library/math.html for the full list. Below some examples to illustrate what can be done:

```python
import math

# Trigonometry
a = 3.0
b = 4.0
c = math.sqrt(a**2+b**2)
angle = math.asin(a/c)*(180/math.pi)
print('The angle of a right triangle with sides {:4.2f} and {:4.2f} is {:4.1f} degrees.'.format(a,b,angle))

# Logarithms and exponentials
T = 298.15
R = 8.314
print('The free energy for mixing three substances is {:6.3f} kJ/mol.'.format(R*T*math.log(1/3)/1e3))

#Statistics
m = 20
n = 5
p = 4
print('There are {:,d} ways to distribute 5 quanta of energy over 20 available molecules'.format(math.comb(m,n)))
print('There are {:,d} different peptides consisting of {:d} different amino acid residues'.format(math.perm(20,p),p))
```
