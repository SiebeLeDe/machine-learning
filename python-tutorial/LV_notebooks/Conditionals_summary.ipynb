# Conditional statement

The if statement is used for conditional execution of the code. Like loops, these are ended with a colon `:` and indentation is used to indicate which piece of code is used if the *conditional expression* evaluates to `True`. We can also provide an `else:` block that is only executed if the *conditional expression* evaluates to `False`.


- **if Statement** Executes a block of code if a condition is True.
```python
if condition:
    # Code to execute if condition is True
```
- **if-else Statement** Executes one block of code if the condition is True, otherwise executes another block.
```python
if condition:
    # Code to execute if condition is True
else:
    # Code to execute if condition is False
```

- **if-elif-else Statement** Checks multiple conditions. Executes the block of the first True condition. You can add as much **elif** statements as you want.

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition2 is True
else:
    # Code to execute if none of the conditions are True
```

A common mistake is to use = instead of ==:
= is for assignment, while == is for comparison.
```python
if x = 10:  # Incorrect
if x == 10:  # Correct
```

Below we give the list of operators, which are used in conditional statements:

|Operator|	Description|	Example|
|---|---|---|
|==	|Equal to: Checks if two values are equal|	x == 5|
|!=	|Not equal to: Checks if two values differ|	x != 5|
|>	|Greater than|	x > 5|
|<	|Less than|	x < 5|
|>=	|Greater than or equal to|	x >= 5|
|<=	|Less than or equal to|	x <= 5|

````python
'''
If statement example
'''

age = 18
if age >= 18:
    print("You are an adult!")
````

````python
'''
If-else statement example
'''
number = 10
if number % 2 == 0: # You can put any expression of the left and on the right side of the operator
    print("Even number")
else:
    print("Odd number")
````

````python
'''
If-elif-else statement example
'''

marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
````

# Break and continue statements

**Break and continue statements** are used in loops, usually with an **if** statement to check a certain condition.

```Break``` terminates the loop  immediately.

- *Important*: in nested loops, break only exits the inner loop

````python
numbers = [1, 2, 3, 4, 5, 6, 7]

for num in numbers:
    if num == 5:
        print("Number 5 found! Breaking the loop.")
        break
    print(num)

print("Loop ended.")
````

Have a look how ```break``` works in a nested loop:

````python
for i in range(3):
    for j in range(3):
        print(f"i={i}, j={j}")
        if j == i:
            print("Breaking out of the inner loop.")
            break
    print("Continuing outer loop...")

print("Program ended.")
````

A ```continue``` statement skips only the current iteration, the loop continues with the next iteration.

See that in the following example ```print(5)``` is skipped.

````python
numbers = [1, 2, 3, 4, 5, 6, 7]

for num in numbers:
    if num == 5:
        continue
    print(num)

print("Loop ended.")
````

**Nested Conditional Statements** 

There are many situations in which you want to check for additional conditions. You can use nested if statements in that case.

````python
n = 8
print(f'n = {n}')
if n % 2 == 0:
   if n % 3 == 0:
      print ("Divisible by 3 and 2")
   else:
      print ("divisible by 2 and not divisible by 3")
else:
   if n % 3 == 0:
      print ("divisible by 3 and not divisible by 2")
   else:
      print ("not Divisible by 2 and not divisible by 3")
````

Another example of usage of nested conditional statements:

````python
x = 10
y = 20
if x < y:
    print("x is less than y")
else:
    if x > y:
        print("x is greater than y")
    else:
        print("x and y are equal")
````

# Using Logical Operators

In many cases you need to check multiple conditions that need to be satisfied. Logical operators allow to combine multiple conditions:

- and (Logical AND): Both conditions must be True.
- or (Logical OR): At least one condition must be True.
- not (Logical NOT): Negates the condition.

|Operator	|Example	|Meaning|
|---|---|---|
|and|	a and b	|True only if both the operands are True|
|or	|a or b	| True if at least one of the operands is True|
|not|	not a|True if the operand is False|

````python
x = 10
y = 20

# Check if both conditions are true
if x > 5 and y < 25:
    print("Both conditions are true.")
else:
    print("At least one condition is false.")
````

````python
x = 10
y = 30

# Check if at least one condition is true
if x > 5 or y < 25:
    print("At least one condition is true.")
else:
    print("Both conditions are false.")
````

````python
x = 10
y = 5

# Check if the condition is NOT true
if not (x < y):
    print("x is not less than y.")
else:
    print("x is less than y.")
````
